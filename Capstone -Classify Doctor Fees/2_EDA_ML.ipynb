```python
#read csv to df
df = pd.read_csv('.\dataset\clean_Final_Train.csv')
```

```python
#Median doctor's experience is 15 years
df.Experience.describe().round()
```

```python
sns.boxplot(x=df['Experience'],data=df);
plt.title('Boxplot of doctor\'s experience');
```

```python
#Variance observed which would be a good categorical variable
plt.figure(figsize=(10, 5))
sns.violinplot(x=df.Profile , y=df.Fees)
plt.title('Fees range for type of doctor');
```

```python
#Cheapest fees are Ayurveda,Dentist
#Most expensive are ENT Specialist, General Medicine and Dermatolgists
#Most expensive starting fees is Dermatologists
```

```python
df.groupby(['Profile','city'])['Fees'].median().to_frame().unstack().plot(kind='bar',stacked=True,figsize=(17, 8))
plt.legend(loc='best')
plt.title('Fees range based on type of doctor');
```

```python
##Dermatologists charges the highest fees
##Ayurveda charges the least
```

```python
df.groupby(['city','Profile'])['Fees'].median().to_frame().unstack().plot(kind='bar',stacked=True,figsize=(17, 8))
plt.legend(loc='best')
plt.title('Fees range based on cities');
```

```python
# Delhi, Bangalore , Hyderabad and Mumbai have one of the highest fees as these are the top tier cities
# These 4 cities are in the top 5 ranking https://en.wikipedia.org/wiki/List_of_cities_in_India_by_population#1_to_25
# The rest are lower tier cities
# We can feature engineer this :)
```

```python
df.groupby(['Experience','Profile'])['Fees'].median().to_frame().unstack().plot(kind='line',stacked=True,figsize=(10, 7))
plt.legend(loc='best')
plt.title('Fees range based on Experience');
# after 37 years experience, medical fees has a downward trend
# fees peak around 22,28,33 years
# 0 to 3 years are the cheapest
```

```python
df.groupby(['Rating','Profile'])['Fees'].median().to_frame().unstack().plot(kind='line',stacked=True,figsize=(10, 7))
plt.legend(loc='best')
plt.title('Fees range based on binned(Rating)');
#Ratings around 80 and above commands higher fees
```
